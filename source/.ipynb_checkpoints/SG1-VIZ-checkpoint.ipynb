```python
def cal_max_min_two_df(df1,df2):
    df1max = np.max(np.max(df1))
    df1min = np.min(np.min(df1))
    df2max = np.max(np.max(df2))
    df2min = np.min(np.min(df2))
    return max(df1max,df2max), min(df1min,df2min)
```
